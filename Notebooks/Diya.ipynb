```python
df = df.drop(columns=['Country Code'])
```

```python
df['2024 [YR2024]'] = pd.to_numeric(df['2024 [YR2024]'], errors='coerce')
df = df.dropna(subset=['2024 [YR2024]'])

df = df.drop(columns=['Series Name'])
df = df.rename(columns={'2024 [YR2024]': 'GDP per capita (PPP)'})
df
```

```python
region_keywords = [
    "Africa", "Asia", "Europe", "Latin America", "Middle East",
    "income", "World", "OECD", "Caribbean"
]

df_regions = df[df['Country Name'].str.contains('|'.join(region_keywords))]

# Reset the index
df_regions = df_regions.reset_index(drop=True)
```
